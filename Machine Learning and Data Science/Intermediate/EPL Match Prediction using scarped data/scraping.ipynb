```python
import requests
```

```python
standings_url="https://fbref.com/en/comps/9/Premier-League-Stats"#replacing the url with a variable
```

```python
data=requests.get(standings_url)#makes req to server and dowloads html of this page
```

```python
 #data.text#shows the html file
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(data.text)#soup object of BeautifulSoup class
```

```python
standings_table=soup.select('table.stats_table')[0]#This is a css selector and its syntax is tag_name.class_name   and [0] coz we only want the first table and returned as a list e.g. Here the seasons table is selected
standings_table
```

```python
links=standings_table.find_all('a')#find all only finds the tags i.e here finds all the a tags in standings_table i.e stats of each and every club e.g each club like chelsea,man city is selected 
```

```python
links=[l.get("href") for l in links]#finds all the href of all the selected as i.e the links of pages of all the clubs e.g puts all the webpages of each clubs in a single list
```

```python
links = [l for l in links if '/squads/' in l]#only select the stats only for squads
links
```

```python
team_urls = [f"https://fbref.com{l}" for l in links]#adds the "https://fbref.com for each element in l
team_urls
```

```python
team_url=team_urls[0]
data = requests.get(team_url)#for 1 team only(here Arsenal)
```

```python
import pandas as pd
matches = pd.read_html(data.text, match="Scores & Fixtures")#match is the one that searches for the string in the tables and returns as a list
```

```python
team_data = matches[0].merge(shooting[["Date", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]], on="Date")#only take the Date,Shots,Shots on Target,Distance(by a shot),Freekicks,Penalty Kicks,Penalty Kicks Attempted and merges based in Date
#matches and shooting dataframes get merged
```

```python
years=[]
no_of_seasons=int(input("Enter the no. of seasons"))
for i in range(no_of_seasons):
    years.append(2023-i)
years
```

```python
#last scraping was done for only one team and one season,now for all teams and the selected seasons
```

```python
all_matches=[]#main list that would be a list of data frames keeping all the merged tables for each and every team
import time
for year in years:
    data=requests.get(standings_url)
    soup=BeautifulSoup(data.text)
    standings_table = soup.select('table.stats_table')[0]
    
    links = [l.get("href") for l in standings_table.find_all('a')]
    links = [l for l in links if '/squads/' in l]
    team_urls = [f"https://fbref.com{l}" for l in links]#making the relative links absolute
    
    previous_season = soup.select("a.prev")[0].get("href")
    standings_url = f"https://fbref.com{previous_season}"
    
    for team_url in team_urls:
        team_name=team_url.split('/')[-1].replace("Stats","").replace("-"," ")#making team-name-stats to team name
        data = requests.get(team_url)#now for multiple teams
        matches = pd.read_html(data.text, match="Scores & Fixtures")[0]
        soup = BeautifulSoup(data.text)
        links = [l.get("href") for l in soup.find_all('a')]
        links = [l for l in links if l and 'all_comps/shooting/' in l]#select shooting from all_comps
        data = requests.get(f"https://fbref.com{links[0]}")#making the relative links absolute
        shooting = pd.read_html(data.text, match="Shooting")[0]
        shooting.columns = shooting.columns.droplevel()
        try:
            team_data = matches.merge(shooting[["Date", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]], on="Date")
        except ValueError:#trying to merge data frames if gives value error as for some teams data not present 
            continue
        team_data = team_data[team_data["Comp"] == "Premier League"]#only keeping Premeir League matches
        team_data["Season"] = year
        team_data["Team"] = team_name
        all_matches.append(team_data)
        time.sleep(5)#done so that the website doesnt block us
```

```python
len(all_matches)
```

```python
match_df = pd.concat(all_matches)#concatinating the list of dataframes into a single dataframe
```

```python
match_df
```

```python
match_df.columns = [c.lower() for c in match_df.columns]#for ease making all headings in lowercase
```
